```python
def knn(k,dist):
  if dist == 'euclidean':
    p = 2

  elif dist == 'manhattan':
    p = 1
  elif dist == 'minkowski':
    p = int(input('Enter value of p:- '))

  top_k_ind =  (np.sum(np.abs(X-X_sample)**p,axis = 1)**(1/p)).sort_values().index[:k]

  output = y[top_k_ind].mode()[0]
  return output
```

```python
knn(3,'minkowski')
```

```python
class Knn:
  def __init__(self,k,method):
    self.k = k
    self.method = method

  def fit(self,x,y):
    self.x = x
    self.y = y

  def predict(self,x_sample):
    if self.method == 'euclidean':
      p = 2

    elif self.method == 'manhattan':
     p = 1
    elif self.method == 'minkowski':
      p = int(input('Enter value of p:- '))

    top_k_ind =  (np.sum(np.abs(self.x-x_sample)**p,axis = 1)**(1/p)).sort_values().index[:self.k]

    output = self.y[top_k_ind].mode()[0]
    return output
```

```python
model = Knn(k = 5, method = 'euclidean')
```

```python
X = df.drop('class',axis = 1)
y = df['class']

#y = f(x)
```

```python
model.fit(X,y)
```

```python
sample = np.array([5.2, 9.4, 2.4, 0.5])
sample
```

```python
model.predict(sample)
```
