```python
import numpy as np

x = np.random.rand(10)
y = 2 * x +1

# error = 1/n(y-(mx+c))^2
# line_eq (y)= mx+c

m = 0
c = 0
lr = 0.01


def derivative(x,y,m,c,lr):
  dldm = 0
  dldc = 0
  N = x.shape[0]
  for xi,yi in zip(x,y):
    yhat  = m*xi+c
    dldm += -2*xi *(yi-yhat)
    dldc += -2*(yi-yhat)
  m = m-lr*(1/N)*dldm
  c = c-lr*(1/N)*dldc
  return (m,c)

epochs = 500
for i in range(epochs):
  N = x.shape[0]
  m,c = derivative(x,y,m,c,lr)
  pred = x * m + c

  loss = (1/N)*np.sum((y-pred)**2, axis = 0)
  print(f'Epoch {i+1}: Loss = {loss:.4f}, m = {m:.4f}, c = {c:.4f}')



derivative(x,y,m,c,lr)
```
